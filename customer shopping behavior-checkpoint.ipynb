```python
import pandas as pd
```

```python
#reading file
df=pd.read_csv(r"C:\Users\aqdas\Downloads\customer_shopping_behavior.csv")
```

```python
#checking file
df.head(20)
```

```python
#cheking any null value
df.isnull().sum()
```

```python
#replacing null values using median of same category
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
#confirming null values
df.isnull().sum()
```

```python
df.columns
```

```python
#arranging column name in lower case and removing spaces
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
#remove (USD) in columns
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
#removing same columns that has no use
df[['discount_applied','promo_code_used']]
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df=df.drop('promo_code_used',axis=1)
```

```python
df.columns
```

```python
#adding frequency purchased days columns for better insights
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df.head(10)
```

```python
#adding age group columns for better analysis
labels=['Young Adult','Adult','Middle Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```
